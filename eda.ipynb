## Exploratory Data Analysis for Midway Checkpoint

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
cities_df = pd.read_csv("data/cleaned_cities500_wide.csv")
ribcot_df = pd.read_csv("data/cleaned_ribcot_2014.csv")
```

```python
# Merge datasets on state
df = pd.merge(cities_df, ribcot_df, on=["state"], how="inner")

df = pd.merge(cities_df, ribcot_df, on=["state"], how="inner")

# ----- Box Plot: Regional Variation in Cardiovascular Disease Cases -----
plt.figure(figsize=(15, 6))
sns.boxplot(x="state", y="cardiovascular", data=df)
plt.xticks(rotation=90)  # Rotate state labels for readability
plt.title("Regional Variation in Cardiovascular Disease Cases")
plt.xlabel("State")
plt.ylabel("Cardiovascular Disease Cases")
plt.show()

```

## Visualization 1
- This figure shows the different cardiovascular disease cases by state. We can do further analysis on this to see the differences between each state.
- We must do some follow up work by comparing the different states and their different values on variables that may explain the differences in cardiovascular disease cases.

```python
df = df.select_dtypes(include = "number")
correlation = df.corr()["cardiovascular"].drop("cardiovascular").sort_values(ascending=False)

# Plot correlation values as a bar chart
plt.figure(figsize=(10, 6))
correlation.plot(kind="bar", color="skyblue", edgecolor="black")

# Formatting the plot
plt.xlabel("Features")
plt.ylabel("Correlation with Cardiovascular Disease")
plt.title("Feature Correlation with Cardiovascular Disease")
plt.xticks(rotation=45, ha="right")

plt.show()
```

## Visualization 2

- We can take away from this visualization that obesity and diabetes is one of the biggest indicators of cardiovascular disease.
